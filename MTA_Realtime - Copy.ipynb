```python
#Connecting to MTA Realtime Data Feeds
#Using General Transit Feed Specification (GTFS) Realtime Protocol Buffers

import requests
import nyct_realtime_pb2
import gtfs_realtime_pb2
import UnixArrivalTime
import TrainIdentifier

onetoseven_url = "https://api-endpoint.mta.info/Dataservice/mtagtfsfeeds/nyct%2Fgtfs"

ace = onetoseven_url + "-ace"
bdfm = onetoseven_url + "-bdfm"
g = onetoseven_url + "-g"
jz = onetoseven_url + "-jz"
nqrw = onetoseven_url + "-nqrw"
l = onetoseven_url + "-l"

#numerical url is not included in this array ....?
API_list = [ace, bdfm, g, jz, nqrw, l]
```

```python
#API_list is not brought into this function yet it is called there. same with onetoseven_url.
def get_correct_url(name):
    if isinstance(name, str):
        name = name.lower()
    else:
        return onetoseven_url

    for api in API_list:
        if name in api:
            return api

def get_train_info(name):
    
    url = get_correct_url(name)
    response = requests.get(url)

    if response.status_code == 200:
        data = response.content
        print(f"Data for {name.upper()} train retrieved successfully.")
        # Further processing of data can be done here
        return data
    else:
        print(f"Failed to retrieve data for {name.upper()} train. Status code: {response.status_code}")


train_name = "a"
a_train_data = get_train_info(train_name)
#So far we have just retrieved the data from the MTA API for the ACE trains... All of them, with stops not specified and there are several of each train.
```

```python
#Parsing GTFS Realtime Data
from datetime import datetime
from zoneinfo import ZoneInfo
acefeed = gtfs_realtime_pb2.FeedMessage()
acefeed.ParseFromString(a_train_data)


print(round(datetime.now().timestamp()))
print(acefeed.header)


for entity in acefeed.entity:
    if entity.HasField('trip_update'):
        for update in entity.trip_update.stop_time_update:
            if update.stop_id == "A43N":
                #print(update)
                print(f"A Manhattan bound {entity.trip_update.trip.route_id} Train is arriving at Lafayette in {UnixArrivalTime.time_until_arrival(update.arrival.time)} minutes.")
            #elif update.stop_id == "A43S":
                #print(f"A Brooklyn bound {entity.trip_update.trip.route_id} Train is arriving at Lafayette in {UnixArrivalTime.time_until_arrival(update.arrival.time)} minutes.")
        
```
