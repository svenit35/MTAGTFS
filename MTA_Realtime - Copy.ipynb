```python
#Connecting to MTA Realtime Data Feeds
#Using General Transit Feed Specification (GTFS) Realtime Protocol Buffers

import requests
import nyct_realtime_pb2
import gtfs_realtime_pb2
import UnixArrivalTime
import TrainIdentifier

onetoseven_url = "https://api-endpoint.mta.info/Dataservice/mtagtfsfeeds/nyct%2Fgtfs"
ace = onetoseven_url + "-ace"
bdfm = onetoseven_url + "-bdfm"
g = onetoseven_url + "-g"
jz = onetoseven_url + "-jz"
nqrw = onetoseven_url + "-nqrw"
l = onetoseven_url + "-l"

url_dict = {
    "ace": ace,
    "bdfm": bdfm,
    "g": g,
    "jz": jz,
    "nqrw": nqrw,
    "l": l,
    "1234567" : onetoseven_url
}
```

```python
#Creates a dictionary of feeds at the call time for each train line. Keys are same keys as url_dict.
dict_of_feeds = {}
for key, value in url_dict.items():
    response = requests.get(value)
    if response.status_code == 200:
        data = response.content
        print(f"Data for {key.upper()} train(s) retrieved successfully.")
        x = gtfs_realtime_pb2.FeedMessage()
        x.ParseFromString(data)
        l = x.entity
        dict_of_feeds[key] = l
    else:
        print(f"Failed to retrieve data for {key.upper()} train(s). Status code: {response.status_code}")
```

```python
#When given a character / string of one of the train names (e.g. "a" for A train), returns the API url of that train from url_dict. Trains are grouped together, so "a" will also return "c" and "e" train data. 
def get_correct_url(name):
    if isinstance(name, str):
        name = name.lower()
    else:
        return onetoseven_url

    for key, value in url_dict.items():
        if name in key:
            return value

#When given a character / string of one of the train names (e.g. "a" for A train), requests the data stream from the API url of that train. Trains are grouped together, so "a" will also return "c" and "e" train data. 
def get_train_data(name):
    
    url = get_correct_url(name)
    response = requests.get(url)

    if response.status_code == 200:
        data = response.content
        print(f"Data for {name.upper()} train retrieved successfully.")
        # Further processing of data can be done here
        return data
    else:
        print(f"Failed to retrieve data for {name.upper()} train. Status code: {response.status_code}")


train_name = "a"
a_train_data = get_train_data(train_name)
#So far we have just retrieved the data from the MTA API for the ACE trains... All of them, with stops not specified and there are several of each train.
```
